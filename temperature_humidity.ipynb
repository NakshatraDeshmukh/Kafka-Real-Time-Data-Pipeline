```python
from pymongo import MongoClient
import pandas as pd

# Connecting to MongoDB
client = MongoClient('localhost', 27017)  
db = client['real_time_data']  
collection = db['sensor_data']  

# Retrieving the data from MongoDB and load it into a DataFrame
data = pd.DataFrame(list(collection.find()))  # Fetching all documents in the collection and convert them to a DataFrame

# Displaying the first few rows of the data
print(data.head()) 
```

```python
# Checking for any missing values in the DataFrame
print(data.columns)
print(data.isnull().sum())  # Displays number of missing values in each column
```

```python
import matplotlib.pyplot as plt

# Creating a plot for Temperature and Humidity
plt.figure(figsize=(10, 6))  # Seting the size of the plot (width, height)

# Ploting the Temperature data
plt.plot(data['temperature'], label='Temperature', color='blue')  # Line plot for temperature

# Ploting the Humidity data
plt.plot(data['humidity'], label='Humidity', color='green')  # Line plot for humidity

# Customizing the plot (labels, title, and legend)
plt.title('Temperature and Humidity Over Time')  # Plot title
plt.xlabel('Time')  # x-axis
plt.ylabel('Value')  # y-axis

# Add a legend to differentiate between temperature and humidity

plt.legend()

plt.grid(True)  # grid lines

plt.savefig('temperature_humidity_plot.png')

# Displaying the plot
plt.show()  
```

```python
# Creating two subplots 
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))  # 2 rows, 1 column

# Ploting temperature on the first subplot
ax1.plot(data['temperature'], label='Temperature', color='blue')
ax1.set_title('Temperature Over Time')
ax1.set_xlabel('Time')
ax1.set_ylabel('Temperature')
ax1.grid(True)
ax1.legend()

# Ploting humidity on the second subplot
ax2.plot(data['humidity'], label='Humidity', color='green')
ax2.set_title('Humidity Over Time')
ax2.set_xlabel('Time')
ax2.set_ylabel('Humidity')
ax2.grid(True)
ax2.legend()

plt.savefig('temperature_humidity_subplots.png')

plt.tight_layout()  # Automatic adjustments to prevent overlap 
plt.show()
```
